
## 2. Data Cleaning

```python
df['retail_price']=df['retail_price'].fillna(np.mean(df['retail_price']))  #fill with mean
```

```python
df['discounted_price']=df['discounted_price'].fillna(np.mean(df['discounted_price']))#fill with mean
```

```python
df['brand']=df['brand'].fillna('unknown')
```

```python
df['image']=df['image'].fillna('not available')
```

```python
df['description']=df['description'].fillna('not available')
```

```python
df['product_specifications']=df['product_specifications'].fillna('not available')
```

```python
df.info()
```

```python
df.dropna(subset=['uniq_id', 'crawl_timestamp', 'product_url', 'product_name', 'product_category_tree',  'pid', 
    'product_rating',  'overall_rating'], inplace=True)     # droped the missing row
```

## 3. Data Formatting

```python
df['crawl_timestamp']=pd.to_datetime(df['crawl_timestamp']) # change into date format
```

```python
df['retail_price'] = df['retail_price'].astype(int)
df['discounted_price'] = df['discounted_price'].astype(int)  #changed into integer datatype
```

```python
df['product_rating'] = df['product_rating'].replace('No rating available', 0).astype(float)
df['overall_rating'] = df['overall_rating'].replace('No rating available', 0).astype(float)  #changed into float datatype
```
